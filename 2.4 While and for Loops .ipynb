While Loops :While statement is the most iteration construct in the language. 

General Format :

                      while <test>    #Loop test
                            <statements1> #Loop body
                       else              #Optional else
                            <statements>  # Run if didn't exit loop with break
                            
Examples 


```python
x = 'spam'
while x:             #While x is not empty 
    print(x, end = '')
    x=x[1:]           #strip first character off x
    
```

```python
a=0; b=10
while a<b :           # One way to code counter loops
    print (a, end='')
    a+=1              # Or, a=a+1

```

Break, continue , pass the Loop else:

break :
     Jumps out of the closest enclosing loop (past the entire loop statement)

pass 
   
   Does nothing at all : it's empty statement placeholder
   
Loop else block 

Runs if and only if the loop is exited normally (i. e without hitting a break)


General Loop Format 
 
              while <test1>:

2. Continue :- 
 The continue statement causes an immediate jump to the top of a loop. 

```python
x=10
while x:
    x=x-1          #"or,x=-1"
    if x%2!=0:     # Odd? -- skip print
     continue
    print(x, end= '')
```

Because continue jumps to the top of the loop, you dont need to nest the print statement inside an if test; the print is only reached if the continue is not run. If this sounds
similar to a goto in other languages, it should. Python has no goto statement, but
because continue lets you jump about in a program.

When youre first getting started with Python.
For instance, the last example might be clearer if the print were nested under the if:

```python
x=10
while x: 
    x=x-1
    if x% 2== 0:         # even - print 
        print (x, end ='')
```

3. Break :

The break statement causes an immediate exit from a loop. Because the code that follows it in the loop is not executed if the break is reached, you can also sometimes avoid nesting by including a break.


```python
while True :
    name = input ('Enter name:')
    if name == 'stop': break
    age = input('Enter age:')
    print ('Hello', name , '=>', int(age)**2)
```

for Loops:

The for loop is a generic sequence iterator in Python: it can step through the items in
any ordered sequence object.

The for statement works on strings, lists, tuples, other built-in iterables, and new objects that well see how to create later with classes. 

General Format :
      
                         for <target> in <object>:       #Assign object items to target
                           <statements>                  # Repeated loop body: use target
                          else:
                            <statements >                # If we didn't hit a 'break'

With Break and continue statement:
                           
                            for <target> in <object> : # Assign object items to target 
                            <statements>
                            if <test> : break # Exit loop now, skip else
                            if <test> : continue # Go to top of loop now
                            else 
                            <statements>  # If we didn't hit a 'break'

Generating Both Offsets and Items: Enumerate

The enumerate function returns a generator objecta kind of object that supports the
iteration protocol that we will study in the next chapter and will discuss in more detail
in the next part of the book. In short, it has a __next__ method called by the next builtin function, which returns an (index, value) tuple each time through the loop.

```python
S = 'spam'  # traditional Method
offset=0
for item in S:
    print(item, 'appears at offset', offset)
    offset+=1
```

```python
S = 'spam' #Enumerate
for (offset, item) in enumerate(S):
    print(item, 'appears at offset', offset)
```
